##Compare Two Companies

```
dfHDFC = pd.read_csv('hdfc.csv', delimiter = ',')
dfHUL = pd.read_csv('hul.csv', delimiter = ',')
#dfHDFC.head(2)
```

```
df2Com = pd.merge(dfHDFC[['Date','Close Price']],dfHUL[['Date','Close Price']],on='Date')
df2Com.head(2)
```

```
df2Com = df2Com.rename(columns ={'Close Price_x':'HDFC','Close Price_y':'HUL'})
df2Com.head(2)
```

```
startDate = df2Com['Date'][dfHDFC.index[0]]
endDate = df2Com['Date'][dfHDFC.index[-1]]
print(startDate,endDate)
```

```
# Plotting directly from DataFrames with Pandas
df2Com.plot(
    x='Date', 
    y=['HDFC', 'HUL'],
    kind='line', 
    legend=False, 
    color=['blue','red'],
    figsize=(16,10)
    
)

# The plot is now created, and we use Matplotlib style
# commands to enhance the output.
plt.ylabel('Share Price')
plt.xlabel("Date")
plt.title("HDFC & HUL "+str(startDate)+" till "+str(endDate))
plt.legend()
plt.gca().yaxis.grid(linestyle=':')
```

##Compare Three Companies

```
dfLNTO = pd.read_csv('lt.csv', delimiter = ',')
dfLNTO.head(2)
```

```
df3Com = pd.merge(df2Com[['Date','HDFC','HUL']],dfLNTO[['Date','Close Price']],on='Date')
df3Com.head(2)
```

```
df3Com = df3Com.rename(columns ={'Close Price':'L&T'})
df3Com.head(2)
```

```
# Plotting directly from DataFrames with Pandas
df3Com.plot(
    x='Date', 
    y=['HDFC', 'HUL', 'L&T'],
    kind='line', 
    legend=False, 
    color=['blue','red', 'green'],
    figsize=(16,10)
)

# The plot is now created, and we use Matplotlib style
# commands to enhance the output.
plt.ylabel('Share Price')
plt.xlabel("Date")
plt.title("HDFC, HUL, L&T "+str(startDate)+" till "+str(endDate))
plt.legend()
plt.gca().yaxis.grid(linestyle=':')
```

##Scaling the Data

```
df3Com_Scaled = df3Com.copy(deep=True)
df3Com_Scaled['HDFC'] = df3Com_Scaled['HDFC']/df3Com_Scaled.iloc[0]['HDFC']
df3Com_Scaled['HUL'] = df3Com_Scaled['HUL']/df3Com_Scaled.iloc[0]['HUL']
df3Com_Scaled['L&T'] = df3Com_Scaled['L&T']/df3Com_Scaled.iloc[0]['L&T']
df3Com_Scaled.head()
```

```
# Plotting directly from DataFrames with Pandas
df3Com_Scaled.plot(
    x='Date', 
    y=['HDFC', 'HUL', 'L&T'],
    kind='line', 
    legend=False, 
    color=['blue','red', 'green'],
    figsize=(16,10)
)

# The plot is now created, and we use Matplotlib style
# commands to enhance the output.
plt.ylabel('Scaled Share Price')
plt.xlabel("Date")
plt.title("Scaled HDFC, HUL, L&T "+str(startDate)+" till "+str(endDate))
plt.legend()
plt.gca().yaxis.grid(linestyle=':')
```

#Simple Moving Average

```
df = pd.read_csv('hul.csv', delimiter = ',')
df.head(2)
```

```
#df.columns
dropCols = ['Symbol', 'Series',  'Prev Close', 'Last Price',  'Average Price', 'Turnover', 'No. of Trades']
df.drop(dropCols, axis=1, inplace=True)
df.head(3)
```

```
df.rename(columns={'Open Price':'Open','High Price':'Hign','Low Price':'Low','Close Price':'Close','Total Traded Quantity':'Volume'}, inplace=True)
df.head(3)
```

```
df.set_index('Date',inplace=True)
df.head(3)
```

```
df.plot(y='Close', figsize=(16,10))

# The plot is now created, and we use Matplotlib style
# commands to enhance the output.
plt.ylabel('Share Price')
plt.xlabel("Date")
plt.title("Technical Analysis")
plt.legend(loc='upper left')
```

```
df['SMA10'] = df['Close'].rolling(10).mean()
df['SMA30'] = df['Close'].rolling(30).mean()
#df.tail()
#df.head(40)
df.count()
```

```
df.plot(y=['Close','SMA10','SMA30'], 
        figsize=(16,10))

# The plot is now created, and we use Matplotlib style
# commands to enhance the output.
plt.ylabel('Share Price')
plt.xlabel("Date")
plt.title("Technical Analysis")
plt.legend(loc='upper left')
```

#Technical Analysis

```
df = pd.read_csv('bharti.csv', delimiter = ',')
df.head(2)
```

```
keepCols = ['Date','Open Price','High Price','Low Price','Close Price','Total Traded Quantity','No. of Trades']
df = df[keepCols].copy(deep=True)
df.head()
```

```
newNames={'Open Price':'Open','High Price':'High','Low Price':'Low','Close Price':'Close','Total Traded Quantity':'Volume','No. of Trades':'Trades'}
df.rename(columns=newNames, inplace=True)
df.head(3)
```

```
# We need to change the data to pandas datetime format to use the MPL Finance package
# https://stackoverflow.com/questions/26763344/convert-pandas-column-to-datetime
# date time formats at https://strftime.org/
df['Date'] =  pd.to_datetime(df['Date'], format='%d-%b-%Y')
df.head(3)
```

```
df.set_index('Date',inplace=True)
df.head(3)
```
